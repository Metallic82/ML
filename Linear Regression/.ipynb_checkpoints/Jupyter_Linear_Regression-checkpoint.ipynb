# Linear Regression with One Variable

```python
#Data Preprocessing 

#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

import os 
os.chdir("C:\\Users\\kritanu\\Desktop\\Machine Learning A-Z Template Folder\\Part 2 - Regression\\Section 4 - Simple Linear Regression")
os.getcwd()
```

```python
#importing the dataset
dataset = pd.read_csv("Salary_Data.csv")

X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:, 1].values
```

```python
#splitting dataset into training and test
from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3, random_state = 0)
```

```python
#Fitting simple linear regression in the training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
#Predicting the test set results
Y_pred = regressor.predict(X_test)
```

```python
#Visulaising the Training Set results
from pylab import rcParams
rcParams['figure.figsize'] = 15, 10
plt.scatter(X_train, Y_train, marker = 'x', color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title("Salary vs Experience (Training Set)", fontsize = 'large')
plt.xlabel("Years of Experience", fontsize = 'large')
plt.ylabel("Salary", fontsize = 'large')
plt.show()
```

```python
#Predicting the Test set results
from pylab import rcParams
rcParams['figure.figsize'] = 15, 10
plt.scatter(X_test, Y_test, marker = 'x', color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title("Salary vs Experience (Test Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
print("Accuracy = ", regressor.score(X_test, Y_test))
```
